```python
class Chocolate:
    def __init__(self, price: float, additive: str = None):
        if price <= 0:
            raise ValueError("Цена должна быть больше нуля")
        self.price = price
        self.additive = additive

    def show_my_chocolate(self):
            if self.additive:
                print(f"Шоколадный батончик, добавка: {self.additive}")
            else:
                print("Обычный шоколадный батончик")

    def __str__(self):
        return f"Шоколадный батончик, {self.price} рублей"

    print("Создание нескольких объектов с разными аргументами")
try:
    #
    choco1 = Chocolate(50.0, "орехи")
    choco2 = Chocolate(35.5)
    choco3 = Chocolate(75.0, "кокос")

    print("\nТестирование метода show_my_chocolate:")
    choco1.show_my_chocolate()
    choco2.show_my_chocolate()
    choco3.show_my_chocolate()

    print("\nТестирование метода __str__ через print:")
    print(choco1)
    print(choco2)
    print(choco3)

    print("\nПроверка обработки ошибок:")
    try:
        choco4 = Chocolate(-10.0)
    except ValueError as e:
        print(f"Поймано исключение: {e}")

except Exception as e:
    print(f"Произошла ошибка: {e}")
```
